Impute the Null value with simpleImputer

```python
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(strategy="most_frequent")
df[['Alcohol_Intake']]=imputer.fit_transform(df[['Alcohol_Intake']])
```

```python
df.isnull().sum()
```

```python
#Preprocessed some data in the column to work properly
df['Alcohol_Intake']=df['Alcohol_Intake'].replace('Moderate','Medium')
df['Polycystic_Ovary_Syndrome']=df['Polycystic_Ovary_Syndrome'].replace('No',0)
df['Polycystic_Ovary_Syndrome']=df['Polycystic_Ovary_Syndrome'].replace('Yes',1)
```

```python
#Dropped the columns that are not necessary for the Prediction to simplify the model
df=df.drop(columns=['Health_Insurance','Urban_Rural','Pregnancies','Heart_Rate'])
```

perform Ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder

OE=OrdinalEncoder(categories=[
    ['Male','Female','Other'],#Gender-->Male=0,Female=1,other=2
    ['No','Yes'],#Family_History
    ['Low','Medium','High'],#Physical_Activity
    ['Never', 'Former', 'Current'],#Smoking_Status
    ['Low','Medium','High'],#Alcohol_Intake
    ['Low','Medium','High'],#Stress_Level
    ['No','Yes'],#Hypertension
    ['No','Yes'],#Regular_Checkups
    ['No','Yes'],#Medication_For_Chronic_Conditions
    ['No','Yes']#Thyroid_Condition
])

columns=['Gender','Family_History','Physical_Activity','Smoking_Status','Alcohol_Intake','Stress_Level',
         'Hypertension','Regular_Checkups','Medication_For_Chronic_Conditions','Thyroid_Condition']
encoded_ordinal_data=OE.fit_transform(df[columns])
```

```python
ordinal_df=pd.DataFrame(encoded_ordinal_data,columns=columns)
```

```python
df=pd.concat([ordinal_df,df.drop(columns,axis=1)],axis=1)
```

```python
df
```

```python
df.info()
```

<h1>Label encoding for output value</h1>

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()

df['Diabetes_Status']=LE.fit_transform(df['Diabetes_Status'])
```

```python
df['Polycystic_Ovary_Syndrome']=df['Polycystic_Ovary_Syndrome'].astype(int)
```

<h1>Train the model</h1>

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

X=df.drop('Diabetes_Status',axis=1)
y=df['Diabetes_Status']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=0)
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)

```

```python
X_train_scaled
```

Train the model using logistic regression

```python
model=LogisticRegression()
model.fit(X_train_scaled,y_train)
y_pred_lg=model.predict(X_test_scaled)
acc_score_lg=accuracy_score(y_test,y_pred_lg)
print(f'Accuracy: {accuracy_score(y_test,y_pred_lg)}')
print('Classification Report for logistic regression: \n', classification_report(y_test,y_pred_lg))
```

Train the model using decision tree

```python
from sklearn.tree import DecisionTreeClassifier

dt_model=DecisionTreeClassifier()
dt_model.fit(X_train_scaled,y_train)
y_pred_dt=dt_model.predict(X_test_scaled)
acc_score_dt=accuracy_score(y_test,y_pred_dt)

print(f'Accuracy: {acc_score_dt}')
print('Classification Report for Decision tree: \n', classification_report(y_test,y_pred_dt))
```

Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_model= RandomForestClassifier()
rf_model.fit(X_train_scaled,y_train)
y_pred_rf = rf_model.predict(X_test_scaled)
acc_score_rf=accuracy_score(y_test,y_pred_rf)
print('Accuracy: ', accuracy_score(y_test,y_pred_rf))
print('Classification Report for Random forest: \n', classification_report(y_test,y_pred_rf))
```

```python
y_pred_lg,y_pred_dt,y_pred_rf
```

```python
results={
    "Model":["Random Forest", "Decision Tree", "Logistic Regression"],
    "Accuracy":[acc_score_rf,acc_score_dt,acc_score_lg]
}
```

```python
results_df=pd.DataFrame(results)
results_df
```

Visualize the Model Comparison with a Bar Chart

```python
plt.figure(figsize=(8,5))

sns.barplot(x='Model',y='Accuracy',data=results_df,palette="viridis")
plt.title("Accuracy Score from the models")
plt.ylim(0,1)
plt.show()
```

```python
print("Random Forest:\n", classification_report(y_test, y_pred_rf))
print("Decision Tree:\n", classification_report(y_test, y_pred_dt))
print("Logistic Regression:\n", classification_report(y_test, y_pred_lg))
```

Compararing Confusion Matrices

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 4))
models=["Random Forest", "Decision Tree", "Logistic Regression"]
preds=[y_pred_lg,y_pred_dt,y_pred_rf]

for i, (model,pred) in enumerate(zip(models,preds)):
  cm=confusion_matrix(y_test,pred)
  sns.heatmap(cm,annot=True,fmt='d',cmap='coolwarm',ax=ax[i])
  ax[i].set_title(f'{model} Confusion Matrix')
  ax[i].set_xlabel("Predicted")
  ax[i].set_ylabel("Actual")

plt.tight_layout()
plt.show()
```
